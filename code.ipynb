```python
import pandas as pd
import numpy as np
import re
```

```python
data = pd.read_excel('Customer Call List.xlsx')
data
```

```python
data.drop_duplicates(inplace=True)
data.drop(columns='Not_Useful_Column', inplace=True)
```

```python
data
```

```python
data.info()
```

```python
data['Last_Name'] = data['Last_Name'].astype(str)
alphabet = r"[A-Za-z]"
for i, name in enumerate(data['Last_Name']):
    cleaned_name = re.sub(r'[^A-Za-z]', '', name)
    data['Last_Name'][i] = cleaned_name
```

```python
data
```

```python
data.replace('N/a', np.nan, inplace=True)
data.dropna(subset=['Phone_Number'], inplace=True)
data.reset_index()
```

```python
print(data['Do_Not_Contact'].unique())
```

```python
data.replace('nan', np.nan, inplace=True)
data.dropna(subset=['Do_Not_Contact'], inplace=True)
```

```python
data.drop(data[(data['Do_Not_Contact']=='Yes') | (data['Do_Not_Contact']=='Y')].index, inplace=True)
```

```python
data
```

```python
data['Do_Not_Contact'] = data['Do_Not_Contact'].apply(lambda x: 'No' if x == 'N' else x)
```

```python
print(data['Do_Not_Contact'].unique())
```

```python
data
```

```python

data['Paying Customer'] = data['Paying Customer'].apply(lambda x: 'Yes' if x == 'Y' else ('No' if x == 'N' else x))
```

```python
data = data.reset_index()
data
```

```python
def split_address(address):
    parts = address.split(',')
    if len(parts) != 2:
        parts = [address, ""]  # Ensure correct list format
    return parts

data['Address'], data['State'] = zip(*data['Address'].apply(split_address))
```

```python
order = ['index', 'CustomerID', 'First_Name', 'Last_Name', 'Phone_Number',
       'Address','State', 'Paying Customer', 'Do_Not_Contact']
data = data.reindex(columns=order)
```
